```python
import pandas as pd
import numpy as np

# 1. 데이터 불러오기
# 다운로드한 파일명과 경로가 맞는지 확인하세요!
df_raw = pd.read_csv('acute_inhalation.xlsx - Data.csv')

print(f"초기 다운로드 데이터 개수 (Record): {df_raw.shape[0]}개")
print(f"초기 컬럼 개수: {df_raw.shape[1]}개")

# 2. SMILES 구조식이 없는 데이터(결측치) 제거
# NTP ICE 데이터셋에서는 'Structure: SMILES' 또는 'SMILES' 컬럼명으로 존재합니다.
smiles_col = 'Structure: SMILES' if 'Structure: SMILES' in df_raw.columns else 'SMILES'

df_clean = df_raw.dropna(subset=[smiles_col]).copy()
print(f"SMILES 결측치 제거 후 데이터 개수: {df_clean.shape[0]}개")
```
